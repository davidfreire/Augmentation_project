```python
#Import the created class
from augment import augmentation_clss
import numpy as np
from matplotlib import pyplot as plt
import skimage.io
```

```python
image = skimage.io.imread('/Users/dfreire/Dropbox/Web/yo1.jpg')
plt.figure(figsize=(6, 6))
plt.imshow(image)
print('Image shape: {}'.format(np.array(image).shape))
```

## Four modes have been specified (more modes can be added to the class)

### Mode: StrongAug

```python
aug = augmentation_clss('StrongAug')
```

```python
w=10
h=10
fig=plt.figure(figsize=(10, 10))
columns = 4
rows = 5
for i in range(1, columns*rows +1):
    img = aug.augment_img(image,p=1) #p decides if this augmentation will be applied. The most common case is p=1 means that we always apply the transformations.
    fig.add_subplot(rows, columns, i)
    plt.imshow(img.astype(np.uint8))
plt.show()
```
